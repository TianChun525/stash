```python
def dtime(input_str):
    try:
        dt = datetime.datetime.strptime(input_str, "%Y-%m-%d %H:%M:%S.%f")
    except ValueError:
        try:
            dt = datetime.datetime.strptime(input_str, "%Y-%m-%d %H:%M:%S")
        except ValueError:
            raise
    return dt
```
