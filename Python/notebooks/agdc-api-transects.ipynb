```python
from datetime import date
import json
import matplotlib.cm as cm
import numpy
import ogr
import pandas
from datacube.api.model import DatasetType, Satellite, BANDS
from datacube.api.query import list_tiles_as_list
from datacube.config import Config
from EOtools.DatasetDrivers.stacked_dataset import StackedDataset
from EOtools import profiles
from EOtools import coordinates
from EOtools import pq_utils
from IDL_functions import bytscl
from IDL_functions import hist_equal
```

Get the `Config()` as it is needed for querying the database.

```python
config = Config()
```

`DatasetType's` of interest.

```python
nbar_ds_type = DatasetType.ARG25
pq_ds_type = DatasetType.PQ25
```

```python
min_date = date(2000, 01, 01)
max_date = date(2008, 12, 31)
```

```python
satellites = [Satellite(i) for i in ['LS5', 'LS7', 'LS8']]
```

The `cell` or `cells` of interest.  We'll just query a single cell.

```python
x_cell = [145]
y_cell = [-34]
```

```python
tiles = list_tiles_as_list(x=x_cell, y=y_cell, acq_min=min_date,
                           acq_max=max_date,
                           satellites=satellites,
                           datasets=nbar_ds_type,
                           database=config.get_db_database(),
                           user=config.get_db_username(),
                           password=config.get_db_password(),
                           host=config.get_db_host(),
                           port=config.get_db_port())
```

```python
print len(tiles)
```

```python
ds = tiles[0]
```

```python
nbar_ds = ds.datasets[nbar_ds_type]
pq_ds = ds.datasets[pq_ds_type]
```

```python
print nbar_ds.path
print pq_ds.path
```

Using the `.path` method to access the full file pathname and setup a `StackedDataset()` for the `nbar` and `pq` `DatasetType's`.

```python
nbar_sd = StackedDataset(nbar_ds.path)
pq_sd = StackedDataset(pq_ds.path)
```

Define a custom `tile` to create an image subset.

```python
UL_map_xy = (145.2003511, -33.3328126)
LR_map_xy = (145.3234148, -33.4558763)
UL_imgx, UL_imgy = coordinates.convert_coordinates(nbar_sd.geotransform, UL_map_xy, to_map=False)
LR_imgx, LR_imgy = coordinates.convert_coordinates(nbar_sd.geotransform, LR_map_xy, to_map=False)
custom_tile = ((UL_imgy, LR_imgy), (UL_imgx, LR_imgx))
shape = (custom_tile[0][1] - custom_tile[0][0], custom_tile[1][1] - custom_tile[1][0])
print "Dimensions of image subset: {}".format(shape)
print "tile envolope ((ystart, yend), (xstart, xend)): {}".format(custom_tile)
```

Get some band information for the current dataset `nbar_ds`.

```python
bands = [nbar_ds.bands.NEAR_INFRARED.value, nbar_ds.bands.RED.value, nbar_ds.bands.GREEN.value]
all_bands = []
all_bnames = []
for band in nbar_ds.bands:
    all_bands.append(band.value)
    all_bnames.append(band.name)
print bands
print all_bands
print all_bnames
```

Now read the subset using the custom `tile` we defined earlier.  The subset will be a false colour image of bands [4,3,2] as [R,G,B].

```python
data = nbar_sd.read_tile(custom_tile, raster_bands=bands)
```

```python
print data.shape
```

```python
scaled_data = numpy.zeros(data.shape, dtype='uint8')
scaled_data[0] = bytscl(data[0], Min=1)
scaled_data[1] = bytscl(data[2], Min=1)
scaled_data[2] = bytscl(data[1], Min=1)
```

```python
imshow(scaled_data.transpose(1, 2, 0), cmap=cm.Greys_r)
```

Create a transect and generate a spectral profile for each of the available bands in the current dataset `nbar_ds`.

```python
points = ((145.2545680, -33.3784236), (145.2814612, -33.4175803))
```

```python
transects = []
idxs = []
xy_start_ends = []
```

```python
for b in all_bands:
    transect, idx, xy_start_end = profiles.arbitrary_profile(nbar_sd, points, b, from_map=True, cubic=True)
    transects.append(transect)
    idxs.append(idx)
    xy_start_ends.append(xy_start_end)
```

```python
pq_mask = pq_utils.extract_pq_flags(pq_transect.reshape(pq_transect.shape[0], 1, 1))
print pq_mask.shape
pq_mask = pq_mask.reshape(pq_mask.shape[0])
print pq_mask.shape
```

Just to demonstrate; If we have any `zeros` then the sum() will not equal the array shape. If it does equal, then there are no masked pixels.

```python
print pq_mask.sum()
if pq_mask.sum() == pq_mask.shape[0]:
    print "No masked pixels therefore no masking needed."
else:
    print "We have masked pixels, therefore masking needs to be applied against the transects."
```

Just to demonstrate, we'll simulate pixels flagged by the pq process and exclude them from the transect. Remember to use `~` to invert the bits of the mask. i.e. `True` becomes `False` and `False` becomes `True`.
