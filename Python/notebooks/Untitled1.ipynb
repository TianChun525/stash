```python
import numpy
import pandas
ts = pandas.Series(numpy.random.randn(1000), index=pandas.date_range('1/1/2000', periods=1000))
ts = ts.cumsum()
ts.plot()
```

```python
ts.values[:] = ts.values[::-1]
ts.plot()
```

```python
ts.values[:] = ts.values[::-1]
df = pandas.DataFrame()
df['forward'] = ts
ts.values[:] = ts.values[::-1]
df['reverse'] = ts
df.plot()
```

```python
p = df['forward'].values
```

```python
print p[0:10]
print numpy.roll(p[0:10], 1)
print numpy.roll(p[0:10], -1)
```

```python
dx0 = df['forward'] - numpy.roll(df['forward'].values, 1)
dx1 = df['forward'] - numpy.roll(df['forward'].values, -1)
peaks = ((dx0 > 0) & (dx1 > 0))
troughs = ((dx0 < 0) & (dx1 < 0))
```

```python
print peaks
```

```python
plot(df.index.values, p)
plot(df.index.values[peaks], p[peaks], 'ro')
```
