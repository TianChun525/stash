```python
fname_nir = '/short/v10/jps547/DcubeDemo/longreach/NBAR_144_-024_NIR.vrt'
fname_red = '/short/v10/jps547/DcubeDemo/longreach/NBAR_144_-024_R.vrt'
fname_green = '/short/v10/jps547/DcubeDemo/longreach/NBAR_144_-024_G.vrt'
nir_stack = StackedDataset(fname_nir)
red_stack = StackedDataset(fname_red)
green_stack = StackedDataset(fname_green)
```

```python
nir_stack.init_yearly_iterator()
yi = nir_stack.get_yearly_iterator()
```

```python
UL_map_xy = (144.2003511, -23.3328126)
LR_map_xy = (144.3234148, -23.4558763)
UL_imgx, UL_imgy = coordinates.convert_coordinates(nir_stack.geotransform, UL_map_xy, to_map=False)
LR_imgx, LR_imgy = coordinates.convert_coordinates(nir_stack.geotransform, LR_map_xy, to_map=False)
custom_tile = ((UL_imgy, LR_imgy), (UL_imgx, LR_imgx))
shape = (custom_tile[0][1] - custom_tile[0][0], custom_tile[1][1] - custom_tile[1][0])
print shape
print custom_tile
```

```python
x_off = []
y_off = []
for feat in lyr:
    ref = feat.GetGeometryRef()
    poly = json.loads(ref.ExportToJson())
    map_coords = poly['coordinates']
    img_coords = coordinates.convert_coordinates(nir_stack.geotransform, map_coords[0], to_map=False)
    img_coords_offset = [(x - roi_tile[1][0], y - roi_tile[0][0]) for (x, y) in img_coords]
    x_off.append([x for (x, y) in img_coords_offset])
    y_off.append([y for (x, y) in img_coords_offset])
```

```python
colours = ['r-', 'g-', 'b-']
imshow(roi_subset.transpose(1,2,0), cmap=cm.Greys_r)
axis('image')
for i in range(len(colours)):
    plot(x_off[i], y_off[i], colours[i])
```
