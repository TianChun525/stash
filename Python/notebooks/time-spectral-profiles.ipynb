```python
from datetime import date
import cPickle as pickle
import json
import matplotlib.cm as cm
import numpy
import ogr
import pandas
from datacube.api.model import DatasetType, Satellite, BANDS
from datacube.api.query import list_tiles_as_list
from datacube.api.query import list_tiles_wkt
from datacube.config import Config
from eotools.DatasetDrivers.stacked_dataset import StackedDataset
from eotools import profiles
from eotools import coordinates
from eotools import pq_utils
from IDL_functions import bytscl
from IDL_functions import hist_equal
```

```python
vec_fname = '/g/data2/v10/testing_ground/jps547/api-transects/LBG-transect_redo.shp'
```

```python
vec_ds = ogr.Open(vec_fname)
layer = vec_ds.GetLayer(0)
```

```python
feat = layer.GetFeature(0)
ref = feat.GetGeometryRef()
```

```python
config = Config()
```

```python
nbar_ds_type = DatasetType.ARG25
pq_ds_type = DatasetType.PQ25
```

```python
satellites = [Satellite(i) for i in ['LS5', 'LS7', 'LS8']]
```

```python
min_date = date(2000, 01, 01)
max_date = date(2008, 12, 31)
years = range(min_date.year, max_date.year)
```

```python
tiles = list_tiles_wkt(ref.ExportToWkt(), satellites=satellites, years=years, datasets=nbar_ds_type,
                       database=config.get_db_database(),
                       user=config.get_db_username(),
                       password=config.get_db_password(),
                       host=config.get_db_host(),
                       port=config.get_db_port())
```

```python
ds = tiles[0]
nbar_ds = ds.datasets[nbar_ds_type]
pq_ds = ds.datasets[pq_ds_type]
bands = [nbar_ds.bands.NEAR_INFRARED.value, nbar_ds.bands.RED.value, nbar_ds.bands.GREEN.value]
```

Define `StackedDataset` classes of the `NBAR` and `PQ` data that allows for co-ordinate conversions and data reading.

```python
nbar_sd = StackedDataset(nbar_ds.path)
pq_sd = StackedDataset(pq_ds.path)
```

```python
extent = layer.GetExtent()
print "Extent: {}".format(extent)
```

Using the extents/envelope of the vector file, create a custom `tile` that we can read the corresponding data for.

```python
UL_map_xy = (extent[0], extent[3])
LR_map_xy = (extent[1], extent[2])
UL_imgx, UL_imgy = coordinates.convert_coordinates(nbar_sd.geotransform, UL_map_xy, to_map=False)
LR_imgx, LR_imgy = coordinates.convert_coordinates(nbar_sd.geotransform, LR_map_xy, to_map=False)
roi_tile = ((UL_imgy, LR_imgy), (UL_imgx, LR_imgx))
print "ROI subset: {}".format(roi_tile)
```

Create an array to hold the array that will display the false colour image `[NIR, RED, GREEN]`

```python
roi_subset = numpy.zeros((3, roi_tile[0][1] - roi_tile[0][0], roi_tile[1][1] - roi_tile[1][0]), dtype='uint8')
roi_subset[0] = bytscl(nbar_sd.read_tile(roi_tile, raster_bands=bands[0]), Min=1)
roi_subset[1] = bytscl(nbar_sd.read_tile(roi_tile, raster_bands=bands[1]), Min=1)
roi_subset[2] = bytscl(nbar_sd.read_tile(roi_tile, raster_bands=bands[2]), Min=1)
```

```python
line = json.loads(ref.ExportToJson())
map_coords = line['coordinates']
```

```python
img_coords = []
for coord in map_coords:
    img_coords.append(coordinates.convert_coordinates(nbar_sd.geotransform, coord, to_map=False))
```

Have a quick look at how complex the transect is. The more vertices the more complex and potentially longer profile compute times. More complex transects should avoid setting the `cubic=True` argument in the `arbitrary_profile` function. Default is `False` in which case `nearest neighbour` interpolation will be used.

```python
print "No. of vertices: {}".format(len(img_coords))
```

```python
img_coords_offset = [(x - roi_tile[1][0], y - roi_tile[0][0]) for (x, y) in img_coords]
x_off = [x for (x, y) in img_coords_offset]
y_off = [y for (x, y) in img_coords_offset]
```

```python
band = nbar_ds.bands.NEAR_INFRARED.value
bname = nbar_ds.bands.NEAR_INFRARED.name
```

Get the profile of the 1st timeslice of the `NIR` band.

```python
transect, idx, xy_start_end = profiles.arbitrary_profile(nbar_sd, img_coords, band)
```

```python
imshow(roi_subset.transpose(1, 2, 0), cmap=cm.Greys_r)
axis('image')
plot(x_off, y_off, 'b-')
```

```python
plot(transect, 'r-')
```

Define the array to hold the `PQ` transects through time.

```python
pq_transects = numpy.zeros((len(tiles), transect.shape[0]), dtype='uint16')
```

Get the corresponding profile for the `PQ` band.

```python
transect, idx, xy_start_end = profiles.arbitrary_profile(pq_sd, img_coords, 1)
```

```python
pq_transects[0] = transect
```

Loop over the rest of the timeslices (ignoring the 1st as it's already been retrieved) and get the profiles for the `PQ` dataset.

```python
for i in range(1, len(tiles)):
    ds = tiles[i]
    pq_sd = StackedDataset(ds.datasets[pq_ds_type].path)
    transect, idx, xy_start_end = profiles.arbitrary_profile(pq_sd, img_coords, 1)
    pq_transects[i] = transect
```

Using the `PQ` profiles mask extract the `ACCA`, `Fmask`, and `Contiguity` bits.

```python
pq_flags = pq_utils.pq_apply_invert_dict()
pq_flags['ACCA'] = True
pq_flags['Fmask'] = True
pq_flags['Contiguity'] = True
```

```python
cloud_mask = pq_utils.extract_pq_flags(pq_transects, flags=pq_flags, combine=True)
```

Also define a mask with all bits selected.

```python
all_mask = pq_utils.extract_pq_flags(pq_transects, combine=True)
```
