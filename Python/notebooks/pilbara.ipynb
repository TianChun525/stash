```python
import pandas
from matplotlib.backends.backend_pdf import PdfPages
```

```python
out_fname = '/g/data/v10/testing_ground/jps547/pilbara/results/plots.pdf'
out_pdf = PdfPages(out_fname)
```

```python
fname = '/g/data/v10/testing_ground/jps547/pilbara/results/pilbara_zonal_stats_results.h5'
store = pandas.HDFStore(fname, 'r')
```

The HDF5 file contains not only the results, but also the attribute table of the vector file. We can see what `groups` are available and get an overview of the data. The `groups` that you should see are `/Metadata`, `/attribute_table`, `/data`.

```python
print store.groups
```

Using the `/attribute_table`, we can create a `join` with the results contained under the `/data` group. This will allow additional queries to be performed.

```python
merge = pandas.merge(store['data'], store['attrbute_table'], how='inner', on='FID') # Forgive the attrbute spelling; it'll be fixed.
```

Set the `DataFrame/table` index to the `Timestamp` column. This allows the `DataFrame/table` to be indexed using `Python datetimes`.

```python
merge.set_index('Timestamp', inplace=True)
```

The results `Table` itself contains various statistics derived from the `Fractional Cover` imagery. For this example we'll only be looking at the column containing the `Mean` values.

```python
print merge.columns
```

We'll perform a simple query and extract the resutls that correspond to the `ex Mt Minnie_Alluvial and River plains` feature type.

```python
g1 = merge[merge['NamAgLanTy'] == 'ex Mt Minnie_Alluvial and River plains']
```

Now we refine the query and retrieve the records for each `Fractional Cover` component.

```python
# Wrong ordering of the fractional components and the FC data will be updated in the next release
npv_g1 = g1[g1['Band'] == 'PV']
pv_g1 = g1[g1['Band'] == 'BS']
bs_g1 = g1[g1['Band'] == 'NPV']
```

`Pandas` has an impressive library of time series techniques availble for the user. The following aggregates the data by `Month`. This is just one of many ways to fill gaps in data.

```python
npv_rs_g1 = npv_g1.resample('M')
pv_rs_g1 = pv_g1.resample('M')
bs_rs_g1 = bs_g1.resample('M')
```

```python
fig, axes = subplots(nrows=1)
bs_rs_g1['Mean'].plot(ax=axes, label='BS')
pv_rs_g1['Mean'].plot(ax=axes, label='PV')
npv_rs_g1['Mean'].plot(ax=axes, label='NPV')
box = axes.get_position()
axes.set_position([box.x0, box.y0 + box.height * 0.1, box.width, box.height * 0.9])
legend(loc='lower left', bbox_to_anchor=(0.1, -0.3), ncol=3)
title('ex Mt Minnie_Alluvial and River plains: Monthly Aggregate')
out_pdf.savefig(fig)
```

This time we'll aggregate the data by `Annual` or yearly.

```python
npv_rs_g1 = npv_g1.resample('A')
pv_rs_g1 = pv_g1.resample('A')
bs_rs_g1 = bs_g1.resample('A')
```

```python
fig, axes = subplots(nrows=1)
bs_rs_g1['Mean'].plot(ax=axes, label='BS')
pv_rs_g1['Mean'].plot(ax=axes, label='PV')
npv_rs_g1['Mean'].plot(ax=axes, label='NPV')
box = axes.get_position()
axes.set_position([box.x0, box.y0 + box.height * 0.1, box.width, box.height * 0.9])
legend(loc='lower left', bbox_to_anchor=(0.1, -0.3), ncol=3)
title('ex Mt Minnie_Alluvial and River plains: Annual Aggregate')
out_pdf.savefig(fig)
```

You can also look at the results for a single polygon. In this example we'll look at `FID 140`. We can also perform the query directly on disk, without having to load the entire results `table` into memory.

```python
df = store.select('data', 'FID == 140')
df.set_index('Timestamp', inplace=True)
```

```python
npv = df[df['Band'] == 'PV']
pv = df[df['Band'] == 'BS']
bs = df[df['Band'] == 'NPV']
```

```python
pv['Mean'].plot(marker='.', color='g', label='PV')
npv['Mean'].plot(marker='.', color='b', label='NPV')
bs['Mean'].plot(marker='.', color='r', label='BS')
title('Fractional Cover Timeseries of FID140')
legend(loc='upper left', bbox_to_anchor=(0, 1.0))
tight_layout()
out_pdf.savefig()
```

```python
npv_rs = npv.resample('A')
pv_rs = pv.resample('A')
bs_rs = bs.resample('A')
```

```python
npv_rs['Mean'].plot(label='NPV', color='b')
pv_rs['Mean'].plot(label='PV', color='g')
bs_rs['Mean'].plot(label='BS', color='r')
title('Fractional Cover Timeseries of FID140: Aggregated Annually')
legend(loc='upper left', bbox_to_anchor=(0, 1.0), prop={'size':8})
tight_layout()
out_pdf.savefig()
```
